```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
data = {'houseId':[],'hours': [],'units':[],'total':[]}
totalEntries = 50
wattage = 8.5
rate = 7.40
```

```python
for i in range(totalEntries):
    data['houseId'].append(i)
    data['hours'].append(random.randrange(100,720))
    data['units'].append(data['hours'][i]* wattage)
    data['total'].append(data['units'][i] * rate) 
```

```python
df = pd.DataFrame(data)
print(data)
```

```python
style.use('ggplot')
plt.plot(data['houseId'],data['units'])
plt.title('Hourly Consumption')
plt.xlabel('houseId')
plt.ylabel('units')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

consumption = ["12  bandra",
          "15  lokhandwala",
          "20  andheri",
          "30  borivali"]

data = [float(x.split()[0]) for x in consumption]
ingredients = [x.split()[-1] for x in consumption]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} )".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Ingredients",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Matplotlib bakery: A pie")

plt.show()
```
